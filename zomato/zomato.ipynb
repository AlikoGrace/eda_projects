```python
# you have to find the mean or average rating of each cusine first 
# Before you are able to find the top 10 because the information there is just for individaual rating
# 1. Group data by cuisine( calculate the aggregate rating for each)
# 2.Sort the results by aggregate rating in descecending order
# 3.Choose the top 10
```

```python
grouped_cuisines=final_df.groupby('Cuisines')['Aggregate rating'].mean().reset_index()
```

```python
sorted_cuisines=grouped_cuisines.sort_values(by='Aggregate rating',ascending=False)
```

```python
sorted_cuisines.head(10)
```

```python
# Finding the top 10 cities by number of restaurants
```
